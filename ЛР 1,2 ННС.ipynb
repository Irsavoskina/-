```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

#Мнения группы экспертов в виде матрицы, лингвистическая переменная "расстояние до города"

      #Терм - Маленькое расстояние SMALL
s = ([[1, 1, 0, 0, 0, 0, 0],     #1 эксперт
     [1, 0, 0, 0, 0, 0, 0],      #2 эксперт
     [1, 0, 0, 0, 0, 0, 0],      #3 эксперт
     [1, 0, 0, 0, 0, 0, 0],      #4 эксперт
     [1, 1, 0, 0, 0, 0, 0],      #5 эксперт
     [1, 1, 1, 0, 0, 0, 0]])     #6 эксперт

      #Терм - Среднее расстояние MIDL
m = ([[0, 0, 1, 1, 0, 0, 0],     #1 эксперт 
     [0, 1, 1, 1, 1, 0, 0],      #2 эксперт 
     [0, 1, 0, 0, 0, 0, 0],      #3 эксперт 
     [0, 1, 1, 1, 0, 0, 0],      #4 эксперт 
     [0, 0, 1, 1, 0, 0, 0],      #5 эксперт 
     [0, 0, 0, 1, 1, 0, 0]])     #6 эксперт 

      #Терм - Большое расстояние BIG
b = ([[0, 0, 0, 0, 1, 1, 1],     #1 эксперт 
     [0, 0, 0, 0, 0, 1, 1],      #2 эксперт 
     [0, 0, 1, 1, 1, 1, 1],      #3 эксперт 
     [0, 0, 0, 0, 1, 1, 1],      #4 эксперт 
     [0, 0, 0, 0, 1, 1, 1],      #5 эксперт 
     [0, 0, 0, 0, 0, 1, 1]])     #6 эксперт 


#Построение ФП на основе статистической обработки мнений группы экспертов
x = [50, 90, 150, 300, 500, 800, 1100]
y = np.mean(s, axis=0)
y1 = np.mean(m, axis=0)
y2 = np.mean(b, axis=0)

plt.figure(figsize=(6, 4))

plt.plot(x, y, marker="o", c="g", label="Маленькое расстояние")
plt.plot(x, y1, marker="o", c="k", label="Среднее расстояние")
plt.plot(x, y2, marker="o", c="r", label="Большое расстояние")
plt.legend(loc=5)
#plt.locator_params(axis="x", nbins=12)
plt.show()
```

```python
#Матрица парных сравнений (один эксперт) терм большое расстояние
mps = ([[1, 1/2, 1/3, 1/5, 1/6, 1/7, 1/9], 
       [2, 1, 1/2, 1/3, 1/5, 1/7, 1/9],
       [3, 2, 1, 1/5, 1/5, 1/7, 1/8],
       [5, 3, 5, 1, 1/3, 1/5, 1/7],
       [6, 5, 5, 3, 1, 1/3, 1/5],
       [7, 7, 7, 5, 3, 1, 1/2],
       [9, 9, 8, 7, 5, 2, 1]])

#Степени принадлежности нечеткого множества (1.10)
mu = 1/(np.sum(mps, axis=0))         # mu = 1/сумму столбцов матрицы
mu_n = mu/max(mu)                    # mu_n субнормальные значения умножаем на максимальное mu

#Построение ФП на основе метода парных сравнений
x = [50, 90, 150, 300, 500, 800, 1100]
y = mu
y1 = mu_n

plt.figure(figsize=(6, 4))

plt.plot(x, y, marker="o", c="r", label="Субнормальное НМ")
plt.plot(x, y1, marker="o", c="g", label="Нормальное НМ")
plt.legend(loc=2)
plt.locator_params(axis="x", nbins=12)
plt.show()
```

```python
#Матрица парных сравнений (один эксперт) терм среднее расстояние
mps2 = ([[1, 1/2, 1/3, 1/6, 1/4, 1/3, 1], 
       [2, 1, 1/2, 1/4, 1/5, 1/2, 1],
       [3, 2, 1, 1, 1/4, 1/2, 1],
       [6, 4, 1, 1, 1/2, 1/2, 1],
       [4, 5, 4, 2, 1, 1/2, 1],
       [3, 2, 2, 2, 2, 1, 1],
       [1, 1, 1, 1, 1, 2, 1]])

#Степени принадлежности нечеткого множества (1.10)
mu2 = 1/(np.sum(mps2, axis=0))         # mu = 1/сумму столбцов матрицы
mu2_n = mu2/max(mu2)                    # mu_n субнормальные значения умножаем на максимальное mu

#Построение ФП на основе метода парных сравнений
x = [50, 90, 150, 300, 500, 800, 1100]
y2 = mu2
y12 = mu2_n

plt.figure(figsize=(6, 4))

plt.plot(x, y2, marker="o", c="r", label="Субнормальное НМ")
plt.plot(x, y12, marker="o", c="g", label="Нормальное НМ")
plt.legend(loc=0)
plt.locator_params(axis="x", nbins=12)
plt.show()
```

```python
#Матрица парных сравнений (один эксперт) терм маленькое расстояние
mps3 = ([[1, 3, 5, 6, 8, 9, 9], 
       [1/3, 1, 3, 5, 6, 7, 8],
       [1/5, 1/3, 1, 4, 5, 6, 7],
       [1/6, 1/5, 1/4, 1, 3, 4, 5],
       [1/8, 1/6, 1/5, 1/3, 1, 3, 4],
       [1/9, 1/7, 1/6, 1/4, 1/3, 1, 1],
       [1/9, 1/8, 1/7, 1/5, 1/4, 1, 1]])

#Степени принадлежности нечеткого множества (1.10)
mu3 = 1/(np.sum(mps3, axis=0))         # mu = 1/сумму столбцов матрицы
mu3_n = mu3/max(mu3)                    # mu_n субнормальные значения умножаем на максимальное mu

#Построение ФП на основе метода парных сравнений
x = [50, 90, 150, 300, 500, 800, 1100]
y3 = mu3
y13 = mu3_n

plt.figure(figsize=(6, 4))

plt.plot(x, y3, marker="o", c="r", label="Субнормальное НМ")
plt.plot(x, y13, marker="o", c="g", label="Нормальное НМ")
plt.legend(loc=1)
plt.locator_params(axis="x", nbins=12)
plt.show()
```

```python
#Степени принадлежности нечеткого множества по координатам собсвтенного вектора(1.10) 
values, vectors = np.linalg.eig(mps)
l_max = values.argmax()
w =  vectors[:, l_max]
w_n =  vectors[:, l_max]/max(w)

values2, vectors2 = np.linalg.eig(mps2)
l2_max = values2.argmax()
w2 =  vectors2[:, l2_max]/sum(vectors2[:, l2_max])
w2_n =  w2/max(w2)

values3, vectors3 = np.linalg.eig(mps3)
l3_max = values3.argmax()
w3 =  vectors3[:, l3_max]/sum(vectors3[:, l3_max])
w3_n = w3/max(w3)

#Построение ФП 
x = [50, 90, 150, 300, 500, 800, 1100]
y = w_n
y1 = w2_n
y2 = w3_n
plt.figure(figsize=(6, 4))

#plt.plot(x, y, marker="o", c="r", label="Субнормальное НМ")
plt.plot(x, y, marker="o", c="g", label="Большое расстояние")
plt.plot(x, y1, marker="o", c="r", label="Среднее расстояние")
plt.plot(x, y2, marker="o", c="k", label="Маленькое расстояние")
plt.legend(loc=0)
#plt.locator_params(axis="x", nbins=12)
plt.show()

print(w3_n)
```

```python
#Разность двух сигмоидных ФП в диапазоне 0, 15
x = np.arange(0, 15.05, 0.1)
mfx = fuzz.dsigmf(x, 2, 4, 11, 14)

#Построение функции принадлежности
plt.figure(figsize=(6, 4))

plt.plot(x, mfx, label="Разность двух сигмоидных ФП")
plt.ylabel("f")
plt.xlabel("x")
plt.ylim(-0.2, 1.1)
plt.legend(loc=4)

plt.show()
```

```python
#Генератор S-образной ФП в диапазоне 0, 10
x = np.arange(0, 10.05, 0.1)
mfx = fuzz.smf(x, 2, 7)

#Построение функции принадлежности
plt.figure(figsize=(6, 4))

plt.plot(x, mfx, "k", label="S-образная ФП")
plt.ylabel("f")
plt.xlabel("x")
plt.ylim(-0.1, 1.1)
plt.legend(loc=4)

plt.show()
```

```python
#Гауcсова нечеткая ФП в диапазоне 0, 20
x = np.arange(0, 20.05, 0.1)
mfx = fuzz.gaussmf(x, 10, 3)

#Построение функции принадлежности
plt.figure(figsize=(6, 4))

plt.plot(x, mfx, "k", label="Гаусcова нечеткая ФП")
plt.ylabel("f")
plt.xlabel("x")
plt.ylim(-0.3, 1.1)
plt.legend(loc=3)

plt.show()
```
